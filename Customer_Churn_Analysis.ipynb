```python
import pandas as pd
from pandas.api.types import is_string_dtype
from sklearn.preprocessing import LabelEncoder,Imputer,StandardScaler
```

```python
data=pd.read_csv('/home/amin/telco-customer-churn (1)/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
data
```

```python
continuous=[]
categorical=[]
for data1 in data:
    if is_string_dtype(data[data1]):
        categorical.append(data1)
    else:
        continuous.append(data1)
```

```python
categorical
```

```python
l1=LabelEncoder()
```

```python
for i in categorical:
    column_change=l1.fit_transform(data[i])
    data=data.drop(columns=[i],axis=1)
    data[i]=column_change
```

```python
data
```

```python
y=data['Churn']
```

```python
y
```

```python
data=data.drop(columns='customerID',axis=1)
```

```python
data=data.drop(columns=['Churn'],axis=1)
```

```python
data
```

```python
impcat=Imputer(missing_values='NaN',strategy='most_frequent',axis=0)
```

```python
impcon=Imputer(missing_values='NaN',strategy='median',axis=0)
```

```python
ddcon=pd.DataFrame(data,columns=continuous)
```

```python
ddcat=pd.DataFrame(data,columns=categorical)
```

```python
ddcat=ddcat.drop(columns='Churn',axis=1)
```

```python
ddcat=ddcat.drop(columns='customerID',axis=1)
```

```python
full_data=pd.concat([ddcat,ddcon],axis=1)
```

```python
full_data_scalar=StandardScaler()
```

```python
full_data_normalized=full_data_scalar.fit_transform(full_data)
```

```python
full_data_new=pd.DataFrame(full_data_normalized,columns=full_data.columns)
```
